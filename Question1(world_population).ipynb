```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#step 8
df = pd.read_csv('/content/Job_1_resource_world_population.csv')
```

```python
# step 10
population_columns = ['2022 Population', '2020 Population', '2015 Population', '2010 Population', '2000 Population', '1990 Population', '1980 Population', '1970 Population']

continent_population_stats = df.groupby('Continent')[population_columns].agg(['mean', 'median', 'var', 'std'])

print(continent_population_stats)
```

```python
df.columns
```

```python
#step 11
year_cols = [col for col in df.columns if "Population" in col]

df["Cumulative_Population"] = df[year_cols].sum(axis=1)
df[["Country/Territory", "Cumulative_Population"]].head()
```

```python
#step 12
new_column_names = {}
for col in df.columns:
    if 'Population' in col and col.replace(' Population', '').isdigit() and len(col.replace(' Population', '')) == 4:
        new_column_names[col] = col.replace(' Population', '')
    elif col == 'World Population Percentage':
        new_column_names[col] = 'World_Population_Percentage'
    elif col == 'Cumulative Population Growth (1970-2022)':
        new_column_names[col] = 'Cumulative_Population_Growth_1970_2022'

df = df.rename(columns=new_column_names)

df.columns
```

```python
#step 13
df.to_csv('processed_data.csv', index=False)
```

```python
# step 14
plt.figure(figsize=(8,5))
sns.histplot(data=df, x="1970", hue="Continent", multiple="stack")
plt.title("Continent-wise Population Distribution in 1970")
plt.savefig('Continent-wise Population Distribution in 1970.png')
plt.show()
```

```python
plt.figure(figsize=(10,5))   # figure width increase
sns.histplot(
    data=df,
    x="1970",
    hue="Continent",
    multiple="stack",
    bins=20
)
plt.title("Continent-wise Population Distribution in 1970")
plt.savefig('Continent-wise Population Distribution in 1970_width.png')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x="2022", hue="Continent", multiple="stack")
plt.title("Continent-wise Population Distribution in 2022")
plt.savefig('Continent-wise Population Distribution in 2022.png')
plt.show()
```

```python
#step 15
year_cols = [str(year) for year in range(1970, 2023) if str(year) in df.columns]
year_cols.sort()

df_melted = df.melt(id_vars=['Continent', 'Country/Territory'], value_vars=year_cols, var_name='Year', value_name='Population')

df_melted['Year'] = df_melted['Year'].astype(int)

continent_population_timeline = df_melted.groupby(['Continent', 'Year'])['Population'].sum().reset_index()

# Plot the timeline of continent-wise population
plt.figure(figsize=(15, 8))
sns.lineplot(data=continent_population_timeline, x='Year', y='Population', hue='Continent', marker='o')
plt.title('Timeline of Continent-Wise Population (1970-2022)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Population', fontsize=12)
plt.ticklabel_format(style='plain', axis='y')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Continent', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('Continent-wise Population Growth Timeline.png')
plt.show()
```

```python
#step 16
corr_matrix = df.select_dtypes(include=np.number).corr()
corr_matrix
```

```python
# Select only numerical columns for correlation calculation
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = df[numerical_cols].corr()

# Display the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Columns', fontsize=16)
plt.savefig('Correlation Matrix.png')
plt.show()
```
